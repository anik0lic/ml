```python
%matplotlib inline
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/drive/')

%cd /content/drive/My Drive/masinsko/
```

```python
def create_feature_matrix(x, nb_features):
    tmp_features = []
    for deg in range(1, nb_features+1):
        tmp_features.append(np.power(x, deg))
    return np.column_stack(tmp_features)

filename = 'bottle.csv'
all_data = pd.read_csv(filename, nrows=700, usecols=['T_degC','Salnty'], dtype='float32')
all_data_wn = all_data.dropna(ignore_index=True)
data = dict()
data['x'] = all_data_wn['T_degC']
data['y'] = all_data_wn['Salnty']

nb_samples = data['x'].shape[0]
indices = np.random.permutation(nb_samples)
data['x'] = data['x'][indices]
data['y'] = data['y'][indices]

data['x'] = (data['x'] - np.mean(data['x'], axis=0)) / np.std(data['x'], axis=0)
data['y'] = (data['y'] - np.mean(data['y'])) / np.std(data['y'])

nb_features = 4
data['x'] = create_feature_matrix(data['x'], nb_features)

plt.scatter(data['x'][:, 0], data['y'])
plt.xlabel('Temperature')
plt.ylabel('Salinity')
plt.show()

w = tf.Variable(tf.zeros(nb_features))
b = tf.Variable(0.0)

learning_rate = 0.01
batch_size = 32
nb_epochs = 100

def pred(x, w, b):
    w_col = tf.reshape(w, (nb_features, 1))
    hyp = tf.add(tf.matmul(x, w_col), b)
    return hyp

def loss(x, y, w, b, reg_lambda):
    prediction = pred(x, w, b)
    y_col = tf.reshape(y, (-1, 1))
    mse = tf.reduce_mean(tf.square(prediction - y_col))
    l2_reg = reg_lambda * tf.reduce_mean(tf.square(w))
    total_loss = mse + l2_reg
    return total_loss

def calc_grad(x, y, w, b, reg_lambda):
    with tf.GradientTape() as tape:
        loss_val = loss(x, y, w, b, reg_lambda)
    w_grad, b_grad = tape.gradient(loss_val, [w, b])
    return w_grad, b_grad, loss_val

adam = tf.keras.optimizers.legacy.Adam(learning_rate=learning_rate)

def train_step(x, y, w, b, reg_lambda):
    w_grad, b_grad, loss_val = calc_grad(x, y, w, b, reg_lambda)
    adam.apply_gradients(zip([w_grad, b_grad], [w, b]))
    return loss_val


lambda_values = [0, 0.001, 0.01, 0.1, 1, 10, 100]
losses = []
for reg_lambda in lambda_values:
    w = tf.Variable(tf.zeros(nb_features))
    b = tf.Variable(0.0)
    for epoch in range(nb_epochs):
        epoch_loss = 0
        for i in range(0, nb_samples, batch_size):
            x_batch = data['x'][i:i+batch_size]
            y_batch = data['y'][i:i+batch_size]
            curr_loss = train_step(x_batch, y_batch, w, b, reg_lambda)
            epoch_loss += curr_loss

        # u svakoj stotoj epohi ispisujemo proseƒçan loss
        epoch_loss /= (nb_samples // batch_size)
        if (epoch + 1) % 10 == 0:
            print(f'Epoch: {epoch+1}/{nb_epochs}| Lambda: {reg_lambda}| Avg loss: {epoch_loss:.5f}')

    # dodavanje finalnog loss-a u listu
    losses.append(epoch_loss)

    print(f'Lambda = {reg_lambda}: w = {w.numpy()}, bias = {b.numpy()}')

    xs = create_feature_matrix(np.linspace(-2, 4, 100, dtype='float32'), nb_features)
    hyp_val = pred(xs, w, b)
    plt.scatter(data['x'][:, 0], data['y'])
    plt.xlabel('Temperature')
    plt.ylabel('Salinity')
    plt.plot(xs[:, 0].tolist(), hyp_val.numpy().tolist(), label=f'Lambda={reg_lambda}', color='y')
    plt.legend()
    plt.show()

plt.plot(lambda_values, losses, marker='o', color='y')
plt.xscale('log')
plt.xlabel('Lambda')
plt.ylabel('Final Loss')
plt.title('Dependence of Final Loss on Lambda')
plt.show()
```
